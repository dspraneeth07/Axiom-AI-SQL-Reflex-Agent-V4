```python
import os
import zipfile
import gdown

# Ensure directory exists
SPIDER_DIR = "data/spider"
os.makedirs(SPIDER_DIR, exist_ok=True)

# Google Drive file ID
file_id = "1403EGqzIDoHMdQF4c9Bkyl7dZLZ5Wt6J"
zip_path = os.path.join(SPIDER_DIR, "spider.zip")

# Download
gdown.download(
    f"https://drive.google.com/uc?id={file_id}",
    zip_path,
    quiet=False
)

# Extract
with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(SPIDER_DIR)

print("‚úÖ Spider dataset downloaded and extracted")
```
